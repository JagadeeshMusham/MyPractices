```python
'''
4) Which was the year when production of Honey in wholeUS was the highest?
'''
```

```python
df3 = df.groupby("year").sum()
df3
```

```python
df3 = df.groupby("year").sum()["production"]
df3
```

```python
df3 = df.groupby("year")["production"].sum()
df3
```

```python
df3 = df.groupby("year")["production"].sum().reset_index().sort_values("production", ascending=False)
df3
```

```python
#Conclusion

'''
The year 2000 was the highest honey produced by all states in US
'''
```

```python
'''
5) From the above inference we get the production was highest in the year 2000, now let
infer which state was having highest contribution in that year
'''
```
