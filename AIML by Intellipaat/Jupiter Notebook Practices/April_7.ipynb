```python
import pandas as pd
```

```python
def read_and_top(path):
    '''
    Function for reading from CSV
    and load the CSV to a data frame
    '''
    df = pd.read_csv(path)
    print(df.head())
    return df
```

```python
titanic = read_and_top("titanic.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(titanic["Fare"])
plt.plot(titanic["Age"])
```

```python
plt.plot(titanic["Fare"], label = "Fare")
plt.plot(titanic["Age"], label = "Age")
plt.legend()
plt.xticks()
```

```python
plt.bar(titanic["Survived"], titanic["Fare"])
```

```python
plt.bar(titanic["Sex"], titanic["Fare"])
```
