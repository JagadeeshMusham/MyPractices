```python
df.drop(columns='customerID', axis=0, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Partner'] = df['Partner'].map({'Yes':1, 'No': 0})
```

```python
df.head()
```

```python
# Label encoder importing
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for index in df.columns:
    if df[index].dtype == 'object':
        df[index] = le.fit_transform(df[index])
```

```python
df.head()
```

```python
# Model Building

X = df.iloc[:, :-1]
y = df['Churn']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 34)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=50, random_state=23) 
```

```python
rf_model
```

```python
rf_model.fit(X_train, y_train) # training the model
```

```python
# predicting or testing the model

y_predict = rf_model.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test, y_predict)
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
nTrees = [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700]
```

```python
for treeCount in nTrees:
    model = RandomForestClassifier(n_estimators=treeCount, random_state=23)
    model.fit(X_train, y_train)
    y_predict = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_predict)
    print("Number of Trees: ", treeCount, " accuracy: ", accuracy)
    
```

```python
# lets use grid search CV(Cross Validation) for further optimazation

param_grid = {
    'n_estimators' : [100, 200, 300],
    'max_depth' :  [None, 5, 10],
    'min_samples_split' : [2, 5, 10],
    'min_samples_leaf' : [1, 2, 4],
    'criterion' : ['gini', 'entropy']
}
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
#performing the random search cv

base_model = RandomForestClassifier(random_state = 23)
grid_search = GridSearchCV(estimator = base_model, param_grid = param_grid, cv = 2)
grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_params_)
```

```python
new_model = RandomForestClassifier(criterion='gini', max_depth= 10, min_samples_leaf=4, min_samples_split=10, n_estimators=300)
```

```python
new_model.fit(X_train, y_train)
```

```python
final_pred = new_model.predict(X_test)
```

```python
accuracy_score(y_test, final_pred)
```

```python
# Print testing summary
print(classification_report(y_test, final_pred))
```
