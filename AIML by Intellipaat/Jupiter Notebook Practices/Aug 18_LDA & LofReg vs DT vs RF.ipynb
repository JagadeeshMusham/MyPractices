```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['bomb_planted'] = le.fit_transform(df['bomb_planted'])
df['map'] = le.fit_transform(df['map'])
df['round_winner'] = le.fit_transform(df['round_winner'])
```

```python
df
```

```python
df.info()
```

```python
#This is another model to encode using iterative method
for column in df.columns:
    if(df[column].dtype == 'object' or df[column].dtype == bool):
        df[column] = le.fit_transform(df[column])
```

```python
X = df.iloc[:, :-1]
y = df['round_winner']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=34)
```

```python
# feature selection
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# LDA - Linear Discriminant Analysis
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(X_train, y_train)
```

```python
lda.transform(X_test)
```

```python
# Coef gives the importance of each column
lda.coef_
```

```python
lda_coefficient = np.exp(np.abs(lda.coef_))
lda_coefficient
```

```python
lda_coefficient = lda_coefficient.flatten()
```

```python
feature_names = X.columns
feature_names
```

```python
plt.figure(figsize = (30,20))

plt.bar(feature_names, lda_coefficient)
plt.xticks(rotation = 90)
plt.xlabel('Feature')
plt.ylabel('score')

plt.show()
```

```python
df_feature_score = pd.DataFrame({'feature_name' : feature_names, 'feature_score' : lda_coefficient} )
```

```python
df_feature_score
```

```python
df_feature_score.sort_values('feature_score')
```

```python
top_20_values = df_feature_score.nlargest(20, 'feature_score')
top_20_values
```

```python
imp_col = top_20_values.index
imp_col
```

```python
X_train
```

```python
X_train = X_train[:, imp_col]
```

```python
X_train
```

```python
X_test = X_test[:, imp_col]
```

```python
X_test
```

```python
pd.DataFrame(X_train).head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import *
```

```python
# Building Logistic Model
logistic_model = LogisticRegression()
```

```python
logistic_model.fit(X_train, y_train)
```

```python
y_pred = logistic_model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)  # accuracy score for Logistic Regression
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Building Decision Tree Model
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(X_train, y_train)
```

```python
y_pred = dt_model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)  # accuracy score for Decision Tree
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Building the Random Forest model
rf_model = RandomForestClassifier()
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred) # accuracy score for Random Forest
```

```python
confusion_matrix(y_test, y_pred)
```
