```python
import numpy as np
import statistics as stats
```

```python
'''
Data of weight of a indian man = 60, 65, 76, 88, 99, 93, 87, 45, 68, 69, 77, 74
sum of all values = 
'''

data = [60, 65, 76, 88, 99, 93, 87, 45, 68, 69, 77, 74]
sum = np.sum(data)
totalValue = np.count_nonzero(data)
totalValue
average = sum / totalValue
print(f"The average of data of weight of a Indian men is: {average}")
```

```python
meanByStats = stats.mean(data)
print(f"The average of data of weight of a Indian men is: {average}")
```

```python
# meadian for the indian men weight by using the numpy library

medianByStats = np.median(data)
print(f"The median value for weight of a indian men is: {medianByStats}")
```

```python
# meadian for the indian men weight by using the stats library

meadianValue = stats.median(data)
print(f"The median value for weight of a indian men is: {meadianValue}")
```

```python
test_scores = [65, 70, 72, 75, 78, 80, 81, 82, 83, 85, 87, 88, 90, 92, 94, 95, 97, 98, 99, 100]

# Calculate percentiles
percentiles = np.percentile(test_scores, [25, 50, 75])

print("25th Percentile (Q1):", percentiles[0])
print("50th Percentile (Median):", percentiles[1])
print("75th Percentile (Q3):", percentiles[2])
```

```python
#how to find the outliers

data = [5.5, 5.6, 6.1, 6.2, 5.9, 5.11, 6.5, 50.5, 0.5, 60.1]

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

IQR = q3-q1

upper_fence = q3 + (1.5*IQR)

lower_fence = q1 - (1.5*IQR)

print(f"Lower fence for the data {data} is: {lower_fence}")
print(f"Upper fence for the data {data} is: {upper_fence}")

for dataPoint in data:
    if dataPoint > upper_fence or dataPoint < lower_fence:
        print(f"Outliers data points are: {dataPoint}")
```

```python
data = [70,75,80, 1, 2,3, 90, 567, 876, 100,105, 110, 125, 130, 120,140, 135, 160, -2, -9]

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

IQR = q3-q1

lowerFence = q1-(1.5*IQR)
upperFence = q3+(1.5*IQR)

print(f"Lower fence for the data {data} is: {lowerFence}")
print(f"Upper fence for the data {data} is: {upperFence}")

for dataPoint in data:
    if dataPoint > upperFence or dataPoint < lowerFence:
        print(f"Outliers data points are: {dataPoint}")
```
